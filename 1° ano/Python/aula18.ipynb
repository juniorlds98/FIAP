União, junção, intersecção e diferença, funciona igual matematica basica 

```python
a = set()
print(a)
print(type(a))
#Colocar itens
a.add(1)
print(a)
a.add(2)
print(a)
a.add("a")
print(a)
a.add([1,2])
print(a)
```

```python
b = [1,2,4,5,6,7]
a = set(b)
c = list(a)
print(a)
print(c)
#Ele apaga automaticamente elementos repetitivos
```

```python
a = set([1,2,3])
b = set([2,3,5])
#Diferença (o que tem em A que não tem em B)
print(a - b)
#União (A junção sem repetir os elementos)
print(a | b)
#Intersecção (O que tem em um que tem em outro)
print(a & b)
```

```python
lista1 = set([1,2,3,4,5])
lista2 = set([4,5,6,7,8])



print(f"A lista tem esses números iguais: {lista1 & lista2}")
print(f"Os valores que só tem na primeira lista: {lista1 - lista2}")
print(f"Os valores que existem apenas na segunda lista: {lista2 - lista1}")
print(f"Os valores que não se repetem: {(lista1-lista2)|(lista2-lista1) }") #Pode ser dado pelo acento circunflexo ^ os elementos não repetidos entre eles


```

Bubble Sort, é o sistema de ordem, você pega a maior bolha e usa ela primeiro, e a menor bolha é sempre o último

```python
def bubble_sort(lista):
    n = len(lista)
    novaLista = lista[:]
    for i in range(n):
        for j in range(n - i - 1):
            if novaLista[j] > novaLista[j+1]:
                temp = novaLista[j]
                novaLista.pop(j)
                novaLista.insert(j+1, temp)        
    return novaLista

numeros = [64,34,25,12,22,11,90]
print("Lista ordenada:", bubble_sort(numeros))
```
