```python
import pandas as pd
import numpy as np
```

```python
#Criando um datagframe ficticio
np.random.seed(42) #Garante a reprodução do experimento. Ele está sorteando 42 vezes tudo aquilo que é random
n = 10000 #Quantidade de clientes que queremos gerar dados ficticios
df = pd.DataFrame({
    "ID": range(1, n + 1),
    "Idade": np.random.randint(18, 65, n),
    "Renda": np.random.randint(2000, 30000, n),
    "Região": np.random.choice(["Norte", "Sul", "Leste", "Oeste"], n)
})
```

```python
df.sample(5)
```

Amostragem aleatoria simples

```python
amostraSimples = df.sample(n = 1000, random_state=42) #random state tem o mesmo papel do seed
amostraSimples.head()
```

Amostragem aleatoria sistematica

```python
intervalo = np.random.randint(1, 50) #Gera intervalo com base na quantidade de registros
amostraSistematica = df.iloc[::intervalo, :]
amostraSistematica.head()
```

Amostragem estratificada

https://scikit-learn.org/stable/

```python
from sklearn.model_selection import train_test_split

amostraEstratificada, _ = train_test_split(df, test_size=0.5, stratify=df["Região"]) #Criando uma estratificação com um dataframe testando uma amostra estratificando por região
amostraEstratificada.head()
```

Amostragem por conglomerado

```python
clusters = df.groupby("Região") #Será uma clusterização por região, ele pegará todos da mesma região
amostraConglomerados = clusters.get_group("Sul")
amostraConglomerados.head()
```

Amostragem por conveniência

```python
amostraConveniencia = df.head(1000) #A convêniencia é sempre por ordenação
amostraConveniencia.head()
```

Amostragem por julgamento

```python
amostraJulgamento = df [(df["Idade"] > 30) &
                        (df["Idade"] <= 55) &
                        (df["Renda"] > 1500) |
                        (df["Região"] != "Oeste")].sample(n = 1000, random_state = 42)
amostraJulgamento.head()
```

Amostragem por cotas

```python
amostraCotas = df.groupby("Região").apply(lambda x: x.sample(n = 25)).reset_index(drop=True) #Esse código está fazendo um sorteio estratificado por região, pegando 25 registros de cada região do dados original, embaralhando aleatoriamente, e guardando em amostraCotas
amostraCotas.head()
```
