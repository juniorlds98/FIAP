Tipos de probabilidade

- Classica (laplace ou teorica): "Qual a chance de tirar um ás de espada do baralho" p = evento ocorre / total eventos possiveis
Evento amostral representada em python: S = {"1", "2", "3", "4", "5", "6"}

- Probabilidade intuitiva: p = eventos similares / total de eventos possiveis

- Probabilidade empirica (frequencia relativa): "Se em 1000 lançamentos de moeda, 480 deram cara, a probabilidade empírica de cara é 48%
P soma de probabilidade desejado / n de vez

- Probabilidade subjetiva baseada em opinião e experiência: "Qual a chance de seu time ganhar o campeonato


Multiplicação: Probabilidade condicional
Soma: probabilidade intuitiva


Teorema de Bayes: Eventos passados na probabilidade do futuro

Probabilidade (Evento A e o Evento B) = P (AxB)/ P(A)

```python
import numpy as np

rng = np.random.default_rng(51)

p_teorica = 1/6
amostras = rng.integers(1, 7, size = 6000)
p_empirica = np.mean(amostras == 6)


print("P(teórica - sair 6):", p_teorica)
print("P(empírica - sair 6):", p_empirica)
print("Obs: probabilidade subjetiva depende de crença/experiência, não de contagem.")
```
